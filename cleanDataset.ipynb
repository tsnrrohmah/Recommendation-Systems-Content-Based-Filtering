```python
import zipfile
import os

# Specify the path to the downloaded zip file
zip_file_path = 'destinasi-wisata-dataset.zip'

# Specify the extraction directory
extracted_path = 'extracted_dataset'

# Create the extraction directory if it doesn't exist
os.makedirs(extracted_path, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)

# List the files in the extraction directory
extracted_files = os.listdir(extracted_path)
print("Extracted Files:", extracted_files)
```

```python
# Melihat dataset wisata
import pandas as pd
data_destinasi = pd.read_csv('/content/extracted_dataset/destinasi-wisata-indonesia.csv')
data_destinasi
```

```python
# Melihat dataset user
data_user = pd.read_csv('/content/extracted_dataset/user.csv')
data_user
```

```python
# Menampilkan informasi tentang dataset, termasuk jumlah nilai yang hilang
print("Informasi Awal Dataset Destinasi Wisata:")
print(data_destinasi.info())

# Menangani nilai yang hilang
# Menghitung jumlah nilai yang hilang untuk setiap kolom
missing_values = data_destinasi.isnull().sum()

# Menampilkan kolom dengan jumlah nilai yang hilang
print("\nJumlah Nilai yang Hilang untuk Setiap Kolom:")
print(missing_values)
```

```python
# Data Cleaning
# Mneghapus kolom yang tidak diperlukan
# menghapus kolom berdasarkan nama
delete_columns =['Time_Minutes', 'Column1', '_1']
data_cleaned = data_destinasi.drop(delete_columns, axis=1)
```

```python
# Menyimpan dataset yang sudah di-cleaned

data_cleaned.to_csv('cleaned_dataset_destinasi.csv', index=False)
```

```python
# Menampilkan informasi setelah data cleaning
print("\nInformasi Dataset Setelah Data Cleaning:")
print(data_cleaned.info())
```
